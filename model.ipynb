```python
cols = ['Camt','Cdur','Htype','Prop','creditScore']
```

```python
df = df[cols]
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['creditScore'] = le.fit_transform(df[['creditScore']])
le_h = LabelEncoder()
df['Htype'] = le_h.fit_transform(df[['Htype']])
le_p = LabelEncoder()
df['Prop'] = le_p.fit_transform(df[['Prop']])
```

```python
from sklearn.preprocessing import StandardScaler
sd = StandardScaler()
import seaborn as sns
sns.boxplot(df['Camt'])
```

```python
import numpy as np
```

```python
per_25 = df['Camt'].quantile(0.25)
per_75 = df['Camt'].quantile(0.75)
iqr = per_75-per_25
df['Camt'] = np.where(df['Camt']>(per_75+(1.5*iqr)),per_75+(1.5*iqr),df['Camt'])
```

```python
sns.boxplot(df['Camt'])
```

```python
df['Camt'] = sd.fit_transform(df[['Camt']])
```

```python
df.sample()
```

```python
x,y = df.drop(columns=['creditScore']),df['creditScore']
```

```python
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm  import SVC
from sklearn.neighbors import KNeighborsClassifier
```

```python
models = [RandomForestClassifier(),GradientBoostingClassifier(),DecisionTreeClassifier(),SVC(),KNeighborsClassifier()]
```

```python
from sklearn.model_selection import cross_val_score
import numpy as np
```

```python
for model in models:
    print(model,end=": ")
    print(np.mean(cross_val_score(model,x,y,cv=5,scoring='accuracy')))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
paramameter= {
    'n_estimators':[300,250,400]

}
```

```python
gd = GridSearchCV(RandomForestClassifier(),param_grid=paramameter)
```

```python
gd.fit(x,y)
```

```python
gd.best_score_
```

```python
gd.best_params_
```

```python
trf1= [('gd',GradientBoostingClassifier(learning_rate=0.2,n_estimators=40)),('scv',SVC()),('rf',RandomForestClassifier(n_estimators=400))]
```

```python
from sklearn.ensemble import VotingClassifier
vc = VotingClassifier(estimators=trf1,n_jobs=-1)
```

```python
np.mean(cross_val_score(vc,x,y,scoring='accuracy',cv=5))
```
